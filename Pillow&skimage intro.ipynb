```python
from skimage import io, color
img = io.imread("C:\\python36\\new_aditya.jpg")
gray = color.rgb2gray(img)
io.imshow(gray)
```

```python
from skimage import data
img = data.astronaut()
img_hsv = color.rgb2hsv(img)
```

```python
io.imshow(img_hsv)
```

```python
img_rgb = color.hsv2rgb(img_hsv)
io.imshow(img_rgb)
```

```python
img_xyz = color.rgb2xyz(img_rgb)
io.imshow(img_xyz)
```

```python
img_cie = color.rgb2rgbcie(img_rgb)
io.imshow(img_cie)
```
